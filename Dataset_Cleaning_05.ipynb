```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CRMLSSold202505_filled .csv')
df.head()
```

```python
cols_to_keep = [
    "MLSAreaMajor",
    "Stories",
    "ViewYN",
    "NewConstructionYN",
    "Levels",
    "PoolPrivateYN",
    "BuyerOfficeAOR",
    "GarageSpaces",
    "LotSizeAcres",
    "LotSizeSquareFeet",
    "LotSizeArea",
    "BuyerOfficeName",
    "BuyerAgentFirstName",
    "BuyerAgentMlsId",
    "StreetNumberNumeric",
    "UnparsedAddress",
    "City",
    "YearBuilt",
    "FireplaceYN",
    "BuyerAgentLastName",
    "LivingArea",
    "Latitude",
    "Longitude",
    "BuyerAgentAOR",
    "BathroomsTotalInteger",
    "ClosePrice",
    "CloseDate",
    "PropertyType",
    "MlsStatus",
    "ParkingTotal",
    "PropertySubType",
    "CountyOrParish",
    "BedroomsTotal",
    "StateOrProvince",
    "PurchaseContractDate",
    "ContractStatusChangeDate",
    "PostalCode"
]

df = df[cols_to_keep]
df.head()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
# Numeric columns with missing values
numeric_cols = [
    "Stories", "GarageSpaces", "LotSizeAcres", "LotSizeSquareFeet",
    "LotSizeArea", "LivingArea", "BedroomsTotal", "BathroomsTotalInteger",
    "YearBuilt", "StreetNumberNumeric", "Latitude", "Longitude", "ParkingTotal"
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
bool_cols = ["NewConstructionYN", "PoolPrivateYN", "ViewYN", "FireplaceYN"]

for col in bool_cols:
    df[col] = df[col].fillna("No")
```

```python

categorical_cols = [
    "MLSAreaMajor", "Levels", "BuyerOfficeAOR", "BuyerOfficeName",
    "BuyerAgentFirstName", "BuyerAgentMlsId", "UnparsedAddress", "City",
    "BuyerAgentLastName", "BuyerAgentAOR", "PropertySubType",
    "CountyOrParish", "StateOrProvince", "PostalCode"
]

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
date_cols = ["PurchaseContractDate", "ContractStatusChangeDate", "CloseDate"]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
yn_cols = ["NewConstructionYN", "PoolPrivateYN", "ViewYN", "FireplaceYN"]

for col in yn_cols:
    df[col] = df[col].map({"Yes": True, "No": False})
```

```python
for col in numeric_cols:  # reuse numeric_cols list from earlier
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
for col in categorical_cols:  # reuse categorical_cols list
    df[col] = df[col].astype("category")
```

```python

df.info()
```

```python
yn_cols = ["ViewYN", "NewConstructionYN", "PoolPrivateYN", "FireplaceYN"]

for col in yn_cols:
    df[col] = df[col].fillna("No").map({"Yes": True, "No": False})
```

```python
df["PropertyType"] = df["PropertyType"].astype("category")
df["MlsStatus"] = df["MlsStatus"].astype("category")
```

```python

df.info()
```

```python

yn_cols = ["ViewYN", "NewConstructionYN", "PoolPrivateYN", "FireplaceYN"]

for col in yn_cols:
    # Normalize: lowercase, strip spaces
    df[col] = df[col].astype(str).str.strip().str.lower()

    # Map cleanly
    df[col] = df[col].map({
        "yes": True, "y": True, "true": True, "1": True,
        "no": False, "n": False, "false": False, "0": False
    })

    # Fill any unmapped values (including NaN) as False
    df[col] = df[col].fillna(False).astype(bool)
```
