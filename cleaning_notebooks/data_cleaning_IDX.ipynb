```python
import pandas as pd
import numpy as np
df_03 = pd.read_csv('CRMLSSold202503_filled.csv')
df_03.head()
```

```python
print(df_03.columns)
print(df_03.shape)
```

```python
def drop_columns(df):
  for col in df.columns:
    if (int(df_03[col].isna().sum()) / df.shape[0]) > 0.50:
      df = df.drop(columns = [col])

  return df
```

```python
filtered_df = df_03[(df_03['PropertyType'] == 'Residential') & (df_03['PropertySubType'] == 'SingleFamilyResidence')]
filtered_df.head()
```

```python
#Columns to Exclude: ListPrice, ListAgentAOR, ListAgentEmail, latfilled, lonfilled, ListingContractDate, ListingID, ListingKeyNumeric, ListAgentFirstName, ListAgentLastName, OriginalListPrice, ListingKey, DaysOnMarket, ListOfficeName, ListAgentFullName, CoListOfficeName, CoListAgentFirstName, CoListAgentLastName, ListingId
```

```python
#filtered_df_2 is the dataframe that excludes the columns we chose to exclude- listing related stuff and ListPrice/OriginalListPrice
filtered_df_2 = filtered_df.drop(columns = ['ListPrice', 'ListAgentAOR', 'ListAgentEmail', 'latfilled', 'lonfilled', 'ListingContractDate', 'ListingId', 'ListingKeyNumeric', 'ListAgentFirstName', 'ListAgentLastName', 'OriginalListPrice', 'ListingKey', 'DaysOnMarket', 'ListOfficeName', 'ListAgentFullName', 'CoListOfficeName', 'CoListAgentFirstName', 'CoListAgentLastName'])
```

##Basic Information about filtered_df_3

```python
filtered_df_3 = drop_columns(filtered_df_2) #filtered_df_3 is the dataframe where I removed all the columns that have 50% or more null values.
```

```python
filtered_df_3
filtered_df_3.isna().sum().sort_values(ascending=False)
```

```python
filtered_df_3.describe()
```

```python
filtered_df_3.info()
```

```python
correlations = filtered_df_3.select_dtypes(include=np.number).corr()['ClosePrice'].abs().sort_values(ascending=False)
correlations
```

```python
numeric_cols = filtered_df_3.select_dtypes(include=np.number).columns.tolist()
categorical_cols = filtered_df_3.select_dtypes(include=['object']).columns.tolist()
bool_cols = ['ViewYN', 'PoolPrivateYN', 'FireplaceYN', 'NewConstructionYN']
for cols in bool_cols:
  categorical_cols.remove(cols)
```

```python
for col in numeric_cols:
  filtered_df_3[col] = filtered_df_3[col].fillna(filtered_df[col].median())

for col in categorical_cols:
  filtered_df_3[col] = filtered_df_3[col].fillna("Unknown")

for col in bool_cols:
  filtered_df_3[col] = filtered_df_3[col].fillna("No")
```

```python
filtered_df_3.info()
#Categorical: BuyerAgentAOR, CloseDate, UnparsedAddress, PropertyType, BuyerOfficename, BuyerAgentMlsId, BuyerAgentFirstName, BuyerAgentLastName, MLSAreaMajor, CountyorParish, MlsStatus, PropertySubType, BuyerOfficeAOR, City, ContractStatusChangeDate, PurchaseContractDate, StateOrProvince
#Boolean: ViewYN, PoolPrivateYN, FireplaceYN, NewConstructionYN,
#Numeric/Continuous: ClosePrice (use median)
#Location: Latitude, Longitude, StreetNumberNumeric, PostalCode (special handling)
#Size: LivingArea, LotSizeAcres, LotSizeArea, LotSizeSquareFeet (use median)
#Count: ParkingTotal, BathroomsTotalInteger, BedroomsTotal, Stories, Levels, GarageSpaces (use zero or mode)
#Time: YearBuilt (median)
```

```python
date_cols = ["PurchaseContractDate", "ContractStatusChangeDate", "CloseDate"]

for col in date_cols:
    filtered_df_3[col] = pd.to_datetime(filtered_df_3[col], errors="coerce")
```

```python
for col in bool_cols:
  filtered_df_3[col] = filtered_df_3[col].replace({'No': False, True: True, False: False})
```

```python
for col in numeric_cols:
  filtered_df_3[col] = pd.to_numeric(filtered_df_3[col], errors='coerce')

for col in categorical_cols:
  filtered_df_3[col] = filtered_df_3[col].astype('category')
```
