```python
import pandas as pd
import numpy as np
```

```python

df_05 = pd.read_csv('CRMLSSold202505_cleaned.csv')
df_04 = pd.read_csv('cleanedCRMLSSold202504.csv')
df_08 = pd.read_csv('cleanedCRMLSSold202508_filled.csv')
df_06 = pd.read_csv('cleanedCRMLSSold202506_filled.csv')
df_03 = pd.read_csv('03_cleaned.csv')
df_02 = pd.read_csv('02_cleaned.csv')
df_07 = pd.read_csv('07_cleaned.csv')
```

```python
df_07 = df_07.drop(columns = ['Unnamed: 0'])
df_02 = df_02.drop(columns = ['Unnamed: 0'])
```

```python
combined_df = pd.concat([df_02, df_03, df_04, df_05, df_06, df_07, df_08], ignore_index=True)
```

```python
# Sanity checks for combined_df
print("Shape:", combined_df.shape)
print("Columns:", combined_df.columns.tolist())
print("\nFirst 5 rows:")
combined_df.head()
```

```python
# Check for missing values
print("Missing values per column:")
combined_df.isnull().sum()
```

```python
# Check data types
print("Data types:")
combined_df.dtypes
```

```python
# Basic statistics for numeric columns
print("Basic statistics:")
combined_df.describe()
```

```python
# Check for duplicate rows
print("Duplicate rows:", combined_df.duplicated().sum())
print("Total rows:", len(combined_df))
print("Unique rows:", len(combined_df.drop_duplicates()))
```

```python
# Check for outliers in key numeric columns (if they exist)
numeric_cols = combined_df.select_dtypes(include=[np.number]).columns
print("Numeric columns:", numeric_cols.tolist())

# Check for extreme values in close price if it exists
if 'ClosePrice' in combined_df.columns:
    print(f"\nClosePrice stats:")
    print(f"Min: ${combined_df['ClosePrice'].min():,.0f}")
    print(f"Max: ${combined_df['ClosePrice'].max():,.0f}")
    print(f"Mean: ${combined_df['ClosePrice'].mean():,.0f}")
    print(f"Median: ${combined_df['ClosePrice'].median():,.0f}")
    
    # Check for zero or negative prices
    zero_prices = (combined_df['ClosePrice'] <= 0).sum()
    print(f"Zero or negative prices: {zero_prices}")
```

```python
# Check data consistency across different time periods
if 'CloseDate' in combined_df.columns:
    print("Date range analysis:")
    combined_df['CloseDate'] = pd.to_datetime(combined_df['CloseDate'], errors='coerce')
    print(f"Date range: {combined_df['CloseDate'].min()} to {combined_df['CloseDate'].max()}")
    
    # Check for future dates or very old dates
    future_dates = (combined_df['CloseDate'] > pd.Timestamp.now()).sum()
    very_old_dates = (combined_df['CloseDate'] < pd.Timestamp('1900-01-01')).sum()
    print(f"Future dates: {future_dates}")
    print(f"Very old dates (<1900): {very_old_dates}")
    
    # Check for invalid dates
    invalid_dates = combined_df['CloseDate'].isnull().sum()
    print(f"Invalid dates: {invalid_dates}")
```

```python
# Check for inconsistent categorical values
categorical_cols = combined_df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols.tolist())

# Check for common data quality issues in categorical columns
for col in categorical_cols[:5]:  # Check first 5 categorical columns
    if combined_df[col].nunique() < 100:  # Only for columns with reasonable number of categories
        print(f"\n{col} unique values:")
        print(combined_df[col].value_counts().head())
        
        # Check for common issues
        empty_strings = (combined_df[col] == '').sum()
        whitespace_only = combined_df[col].str.strip().eq('').sum() if combined_df[col].dtype == 'object' else 0
        print(f"Empty strings: {empty_strings}, Whitespace only: {whitespace_only}")
```
